```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy as sp
```

```python
x_f=np.zeros((64,1),dtype=complex);
for i in range(63,64):
    x_f[i,0]=1+1j
    
x_t= np.fft.fft2(x_f)
```

```python
plt.figure(figsize=(10,8))

ax1=plt.subplot2grid((2,1),(0,0))
ax2=plt.subplot2grid((2,1),(1,0))

ax1.stem(range(64),x_t.real)
ax2.stem(range(64),x_t.imag)
```
